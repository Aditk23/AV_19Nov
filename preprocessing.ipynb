```python
train_new = pd.DataFrame({'Emp_ID':train1.groupby('Emp_ID').groups.keys()})
```

```python
train_new
```

```python
train_new=train_new.set_index('Emp_ID')
train_new
```

```python
train_new['Age']= train1.groupby('Emp_ID').first()['Age']
train_new
```

```python
train_new['Gender']= train1.groupby('Emp_ID').last()['Gender']
train_new
```

```python
train_new['City']= train1.groupby('Emp_ID').last()['City']
train_new['Education_Level']= train1.groupby('Emp_ID').last()['Education_Level']
train_new['Is_promoted']= train1.groupby('Emp_ID').last()['Is_promoted']
train_new['Attrition']= train1.groupby('Emp_ID').last()['Attrition']

train_new
```

```python
train1
```

```python
train_new['Salary']=train.groupby('Emp_ID')['Salary'].agg('mean')
train_new['Total Business Value']=train.groupby('Emp_ID')['Total Business Value'].agg('sum')
train_new['FirstQuaterRating']=train.groupby('Emp_ID').first()['Quarterly Rating']
train_new['LastQuaterRating']=train.groupby('Emp_ID').last()['Quarterly Rating']

```

```python
train_new
```

```python
train_new['Rating_diff'] = train_new['LastQuaterRating'] - train_new['FirstQuaterRating']
train_new
```

```python

def rating(x):
    if(x==0):
        return 'Same'
    elif(x>0):
        return 'Increased'
    else:
        return 'Decreased'
    
    
train_new['RatingChange'] = train_new['Rating_diff'].apply(rating)
```

```python
train_new.RatingChange.value_counts()
```

```python
train_new
```

```python
train_new.drop(['FirstQuaterRating','LastQuaterRating','LastQuaterRating'],axis=1,inplace=True)
train_new
```
